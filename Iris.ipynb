```python
plt.figure(figsize=(8, 6))
plt.scatter(df['sepal_length'], df['sepal_width'], c='blue', label='Iris-setosa')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.show()
```

```python
sns.displot(df,x='sepal_length',y='sepal_width')
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['petal_length'], df['petal_width'], c='green', label='Iris-setosa')
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend()
plt.show()
```

```python
sns.jointplot(df,x='petal_length',y='petal_width')
```

```python
from sklearn.cluster import KMeans
k_rng=range(1,10)
sse=[]
for k in k_rng:
  km=KMeans(n_clusters=k)
  km.fit(df[['petal_length', 'petal_width']])
  sse.append(km.inertia_)
sse
```

```python
plt.xlabel('k_rng')
plt.ylabel("sum of the errors")
plt.plot(k_rng,sse)
```

```python
km=KMeans(n_clusters=3,random_state=0)
y_predicted=km.fit_predict(df[['petal_length','petal_width']])
y_predicted
```

```python
cl = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
cd = df[cl]
correlation_matrix = cd.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix Of Iris Data')
plt.show()
```
